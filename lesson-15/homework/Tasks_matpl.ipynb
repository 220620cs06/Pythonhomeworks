Basic Plotting


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', linewidth=2)
plt.title("Plot of $f(x) = x^2 - 4x + 4$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True)
plt.show()
```

2. Sine and Cosine Plot

```python
x = np.linspace(0, 2*np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y_sin, 'b--', label='sin(x)', marker='o', markersize=4, alpha=0.7)
plt.plot(x, y_cos, 'r-', label='cos(x)', marker='^', markersize=4, alpha=0.7)
plt.title("Sine and Cosine Functions", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

3. Subplots (2x2 Grid)

```python
x = np.linspace(-2, 2, 200)
x_positive = np.linspace(0.01, 5, 200)  # Avoid log(0)

plt.figure(figsize=(10, 8))

# Top-left: Cubic function
plt.subplot(2, 2, 1)
plt.plot(x, x**3, 'r-')
plt.title("$f(x) = x^3$")
plt.xlabel("x")
plt.ylabel("f(x)")

# Top-right: Sine function
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), 'g--')
plt.title("$f(x) = \sin(x)$")
plt.xlabel("x")
plt.ylabel("f(x)")

# Bottom-left: Exponential function
plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x), 'b-.')
plt.title("$f(x) = e^x$")
plt.xlabel("x")
plt.ylabel("f(x)")

# Bottom-right: Logarithmic function
plt.subplot(2, 2, 4)
plt.plot(x_positive, np.log(x_positive), 'm:')
plt.title("$f(x) = \log(x+1)$")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.tight_layout()
plt.show()
```

4. Scatter Plot


```python
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
colors = np.random.rand(100)
sizes = 50 * np.random.rand(100)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')
plt.title("Random Scatter Plot", fontsize=14)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.colorbar(label="Color Intensity")
plt.grid(True)
plt.show()
```

5. Histogram

```python
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Normal Distribution", fontsize=14)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.show()
```

6. 3D Surface Plot


```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title("3D Surface Plot: $f(x, y) = \cos(x^2 + y^2)$", fontsize=14)
ax.set_xlabel("x", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.set_zlabel("f(x, y)", fontsize=12)
plt.show()
```

7. Bar Chart

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['skyblue', 'salmon', 'lightgreen', 'gold', 'violet']

plt.figure(figsize=(8, 5))
plt.bar(products, sales, color=colors, edgecolor='black')
plt.title("Product Sales", fontsize=14)
plt.xlabel("Products", fontsize=12)
plt.ylabel("Sales (units)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

8. Stacked Bar Chart (Categories Over Time)

```python
time_periods = ['T1', 'T2', 'T3', 'T4']
category_a = [20, 35, 30, 35]
category_b = [15, 25, 20, 30]
category_c = [10, 20, 25, 15]

plt.figure(figsize=(8, 5))
plt.bar(time_periods, category_a, label='Category A', color='skyblue')
plt.bar(time_periods, category_b, bottom=category_a, label='Category B', color='salmon')
plt.bar(time_periods, category_c, bottom=np.add(category_a, category_b), label='Category C', color='lightgreen')
plt.title("Stacked Bar Chart: Category Contributions", fontsize=14)
plt.xlabel("Time Periods", fontsize=12)
plt.ylabel("Contribution", fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
