````python
# Find the average age by passenger class
avg_age_by_class = titanic_xls.groupby('Pclass')['Age'].mean()
print("Average age by class:")
print(avg_age_by_class)

# Survival rate by gender
survival_by_gender = titanic_xls.groupby('Sex')['Survived'].mean()
print("\nSurvival rate by gender:")
print(survival_by_gender)

# Survival rate by class
survival_by_class = titanic_xls.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by class:")
print(survival_by_class)
````

````python
# Identify the director with the highest total director_facebook_likes
director_likes = movie_df.groupby('director_name')['director_facebook_likes'].sum()
top_director = director_likes.idxmax()
top_likes = director_likes.max()
print(f"Director with highest total Facebook likes: {top_director} with {top_likes} likes")

# Find the 5 longest movies and their respective directors
longest_movies = movie_df.nlargest(5, 'duration')[['movie_title', 'duration', 'director_name']]
print("\n5 longest movies and their directors:")
print(longest_movies)
````
