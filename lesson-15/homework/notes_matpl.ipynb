plot() - Creates line plots.
scatter() - Creates scatter plots.
bar() - Creates bar charts.
hist() - Creates histograms.
imshow() - Displays images.
show() - Displays the plots.
xlabel() / ylabel() - Labels the x-axis and y-axis.
title() - Adds a title to the plot.
legend() - Adds a legend to the plot.
grid() - Adds grid lines to the plot.
xlim() / ylim() - Sets the limits for the x-axis and y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(50)
b = np.random.randint(0, 50, 50)
sizes = np.abs(np.random.randn(50)) * 100

plt.scatter(a, b, s=sizes, c=b, cmap='cool')
plt.colorbar()
plt.show()
```

```python
categories = ['Red', 'Blue', 'Green', 'Yellow']
values = [10, 15, 7, 12]

plt.bar(categories, values, color=[c.lower() for c in categories], width=0.3)
plt.title('Basic Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
colors = ['red', 'blue', 'green', 'purple']
widths = [0.4, 0.6, 0.8, 1.0]

plt.bar(categories, values, color=colors, width=widths, edgecolor='black')
plt.title('Bar Chart with Custom Colors and Widths')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, c='g')
ax.set_title('3D line plot geeks for geeks')
plt.show()
```

```python
x = np.linspace(-5, 6, 10000)

def f_x(x):
    return 2*x**2 - 3*x + 1

y = f_x(x)

plt.plot(x, y, label='$f(x) = 2x^2 - 3x + 1$')
plt.title('Parabola')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)

xx, yy = np.meshgrid(x, y)

zz = np.sin(np.sqrt(xx**2 + yy**2))

ax.plot_surface(xx, yy, zz, cmap='viridis')
```
